```python
def FDcurve(x, a, b):
    h0 = sqrt(a**2-(b-1)**2)
    F = 2*(1-b+sqrt(a**2-(h0-x)**2))*(h0-x)/sqrt(a**2-(h0-x)**2)
    return F
```

```python
def KSPRING(u, a, b):
    K = -2*(1+(1-b)/sqrt(a**2-u**2)*(1+u**2/(a**2-u**2)))
    return K
```

```python
x = linspace(0,14,1000)
figure(figsize=[5,5]); ax = subplot(111)
ax.plot(x, FDcurve(x, 0.7, 1.1), '-', lw=1, color='red')
ax.plot(x, FDcurve(x, 0.7, 1.2), '-', lw=1, color='blue')
ax.plot(x, FDcurve(x, 0.7, 1.4), '-', lw=1, color='pink')
ax.set_xlim(0,1.4); ax.set_ylim(-1,1)
ax.set_xlabel('x (Displacement)')
ax.set_ylabel('F (Force)')
tight_layout()
```

```python
u = linspace(-0.8,0.8,1000)
figure(figsize=[5,5]); ax = subplot(111)
ax.plot(u, KSPRING(u, 0.7, 1.1), '-', lw=1, color='red')
ax.plot(u, KSPRING(u, 0.7, 1.2), '-', lw=1, color='blue')
ax.plot(u, KSPRING(u, 0.7, 1.4), '-', lw=1, color='pink')
ax.set_xlim(-0.7,0.7); ax.set_ylim(-2,2)
ax.set_xlabel('u (Displacement)')
ax.set_ylabel('K (spring constant)')
tight_layout()
```
